```python
import os
import numpy as np 

from keras.preprocessing.image import ImageDataGenerator
from glob import glob
from keras.preprocessing import image
```

### Config

```python
augTotal = 100 ## total Amount per image 
```

### Build generator

```python
aug = ImageDataGenerator(
	rotation_range=10,
    #brightness_range=[0.5,1.5],
	#zoom_range=0.15, check ratio
	width_shift_range=0.15,
	height_shift_range=0.15,	
	horizontal_flip=True,
	fill_mode="constant")
```

### Load data

```python
ROOT_DIR = os.path.abspath("../../")
DS_DIR = os.path.join(ROOT_DIR, "Datasets/awe_cvledb_aug_100/")
```

```python
for (dirpath, u_dirnames, filenames) in os.walk(DS_DIR):
    dirnames = sorted(u_dirnames)
    
    for i in range(0,len(dirnames)):
        total = 0
        
        FILE_DIR = os.path.join(DS_DIR, str(dirnames[i]))
        files = glob(FILE_DIR+'/*.png') 
        
        dataaug_dir = os.path.join(FILE_DIR, "da")
        if not os.path.exists(dataaug_dir):
            os.mkdir(dataaug_dir)
                
        while total < augTotal:        
            for j in range (0,len(files)):            
                #image_class = int(str(dirnames[i]))-1
                img = image.load_img(files[j])
                img_data = image.img_to_array(img)
                img_data = np.expand_dims(img_data, axis=0)

                imageGen = aug.flow(img_data, batch_size=1, save_to_dir=os.path.join(FILE_DIR, "da/"), save_prefix=str(dirnames[i]), save_format="png")
                for imggen in imageGen:	                
                    break 
                total += 1
        print(str(dirnames[i])+ " done")
    break
```
